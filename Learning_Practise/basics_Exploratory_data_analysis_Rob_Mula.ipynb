#### Shift function

- This function returns the value row concedding or row succedding by manipulating arguments in the shift function

**For Example**

- shift(1) --> Lagged version which means the value of row be succedding towards next row and it iterates over all rows but first row takes null value because there is no above row is present that's why  [Top to Bottom approach] - Lag

- shift(-1) --> Lead version which means the value of row be Concedding towards previous row Becuase it starts with bottom to top and it iterates over all rows but last row takes null value because there is no below row is present that's why [Bottom to Top approach] - Lead

```python
print('Original Df')
display(df1.Speed_Mph)

print()
print('Lagged Version of Df')
display(df1['Speed_Mph'].shift(1))
print()
print('Lead Version of Df')
df1['Speed_Mph'].shift(-1)
```
